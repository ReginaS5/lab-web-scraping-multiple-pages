```python
url = 'https://www.officialcharts.com/charts/afrobeats-chart/'

response = requests.get(url)
response.status_code
```

```python
soup = BeautifulSoup(response.content, "html.parser")
soup
```

```python
#Here starts Sandras very detailed explanation


song_container = soup.find('div', class_='afrobeats-chart-container')
song_container
#If you call your song_container variable you will see it is empty. It is not working because in your page 
#it doesn't exist an html element div that has class 'afrobeats-chart-container'
#You need to use the exact same names for ids or classes that are present on your website
```

```python
'''If you Inspect your webpage and hover over the content you want, you will see each element that contains 
ranking, title of song, artist, picture, etc is a div and all those divs have the same class: chart-item-content
'''
cards = soup.find_all('div', class_='chart-item-content')
cards
```

```python
# Now lets check one element of all that info we got:
cards[0]
'''It has the information we want: Artist name and Title of song, 
but it also has a lot of other extra information we don't want, so we can refine our find_all
so it is more specific.
You need to find the info you want here: thankfully, the artist and song are both inside an
element called audio:
<audio data-artist="TYLA" data-title="WATER" preload="none" src="https://audio-ssl.itunes.apple.com/itunes-assets
/AudioPreview116/v4/33/5d/ed/335ded11-49b3-9a7e-63d7-8ad495ccec05/mzaf_6729827213024194854.plus.aac.p.m4a"
type="audio/mpeg"></audio><div class="action"></div><div class="progress-wrapper">

So we can point to that element and see what we get
'''
```

```python
cards2 = soup.find_all('audio')
cards2
```

```python
#Let's see one element of that result set
cards2[0]
#We now have the data we need and less extra data to throw away
```

```python
#Always develop the code for one case first and after you know it works, you can apply a for loop to all the cases
test1 = cards2[0] #this is our audio element
test1.get("data-artist") #that's how we get the artist!
```

```python
test1.get("data-title") #That's how we get our song title!
```

```python
#Now we can create our for loop: it will take each element of cards2 and get artist and title
#It will store each in a list or a dictionary or as a tuple. That's your choice. I'm doing two lists

artists_list = []
songs_list = []
for card in cards2:
    artist = card.get("data-artist")
    song = card.get("data-title")
    artists_list.append(artist)
    songs_list.append(song)
```

```python
#I print my artists list
print(artists_list)
```

```python
#I print my songs list
print(songs_list)
```

```python
#And I create a dataframe with both lists: https://www.geeksforgeeks.org/create-a-pandas-dataframe-from-lists/
df_afrobeats = pd.DataFrame(zip(artists_list, songs_list), columns=['artist', 'song_title'])
df_afrobeats
```

```python
'''
But we have a problem: We check against the webpage and see we are getting 19 results instead of 10. Why? 
Turns out that number 16: 40 Days by Tamera doesn't have an audio element!
So we need to go back to our website and try another element/class combination
'''
```

```python
#We have a element with class chart-name that includes the song title as text inside a span
cards3 = soup.find_all("a", class_="chart-name")
cards3
```

```python
#One element of a_names. It has several spans and we need to get to the one that contains the song title: WATER
cards3[0]
```

```python
#We iterate through our spans
lst_span = []
for span in cards3[0]:
    lst_span.append(span)
print(lst_span)
```

```python
'''
The info we need is inside the third span:
First span: ' '
Second span: <span class="movement-icon"></span>
Third span: <span>WATER</span>
So we ask for the text of the third span, that has index=2
'''
lst_span[2].get_text() #That's our song title!
```

```python
#Let's do the same with artist and class chart-artist
cards4 = soup.find_all("a", class_="chart-artist")
cards4[0]
```

```python
#We only have one span
cards4[0].span.get_text() #That0s our artist
```

```python
#Now we do this for all the elements
artists_new = []
songs_new= []

#To get all the song titles
for card in cards3:
    lst_span = []
    for span in card:
        lst_span.append(span)
    new_song = lst_span[2].get_text()
    songs_new.append(new_song)

#To get all artists names
for card in cards4:
    new_artist = card.span.get_text()
    artists_new.append(new_artist)
```

```python
print(songs_new)
print(artists_new)
```

```python
#Create your dataframe using both lists
data_afrobeats = pd.DataFrame(zip(artists_new, songs_new), columns=["artist", "song"])
data_afrobeats
```
